```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('sections-sportives-scolaires.csv', sep=';')
```

```python
data.head()
```

```python
df = data.copy()
df["Sections scolaires"] = df["Sections scolaires"].astype(str).str.upper()

mask = (
    df["Sections scolaires"].str.contains("FOOTBALL") |
    df["Sections scolaires"].str.contains("RUGBY")
)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

df2 = df.copy()
df2["Sections scolaires"] = df2["Sections scolaires"].astype(str).str.upper()

mask = df2["Sections scolaires"].str.contains("FOOTBALL") | df2["Sections scolaires"].str.contains("RUGBY")
df_filtered = df2[mask & df2["Région"].isin(["Occitanie", "Hauts-de-France"])]

df_exploded = (
    df_filtered.assign(
        Section_split=df_filtered["Sections scolaires"].str.split(r"[;/\-]+")
    )
    .explode("Section_split")
)

df_exploded["Section_split"] = df_exploded["Section_split"].str.strip()
df_exploded = df_exploded[df_exploded["Section_split"].isin(["FOOTBALL", "RUGBY"])]

section_counts = df_exploded.groupby(["Région", "Section_split"]).size().unstack()

colors = {
    "FOOTBALL": "#1f77b4",
    "RUGBY": "#ff4500",
}

fig, ax = plt.subplots(figsize=(9, 4))

# Centres très rapprochés
indices = np.arange(len(section_counts.index)) * 0.3
height = 0.18  # barres très fines = collement visuel

ax.barh(
    indices,
    section_counts["FOOTBALL"],
    height=height,
    color=colors["FOOTBALL"],
    label="Football"
)

ax.barh(
    indices,
    section_counts["RUGBY"],
    left=section_counts["FOOTBALL"],
    height=height,
    color=colors["RUGBY"],
    label="Rugby"
)

ax.set_yticks(indices)
ax.set_yticklabels(section_counts.index, fontsize=12)

plt.title(
    "Nombre de sections sportives Football & Rugby par région\n"
    "dans l'enseignement secondaire",
    fontsize=14
)

plt.xlabel("Nombre de sections", fontsize=14)
plt.ylabel("Région", fontsize=14)

plt.legend(
    title="Type de section sportive",
    title_fontsize=13,
    fontsize=12,
    loc="lower right",
    bbox_to_anchor=(0, -0.22),
    ncol=2
)

plt.subplots_adjust(bottom=0.28)
plt.tight_layout()
plt.show()
```
